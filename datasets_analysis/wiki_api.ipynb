```python
AUTH_URL = "https://auth.enterprise.wikimedia.com/v1/login"
BASE_URL = "https://api.enterprise.wikimedia.com/v2"

def wiki_login(username, password):
    resp = requests.post(
        AUTH_URL,
        json={"username": username, "password": password}
    )
    resp.raise_for_status()
    return resp.json()["access_token"]


def fetch_wikipedia_article(title: str, token: str):
    headers = {"Authorization": f"Bearer {token}"}
    url = f"{BASE_URL}/pages/enwiki/{title}"
    r = requests.get(url, headers=headers)
    if r.status_code != 200:
        return r.status_code
    return r.json()
```

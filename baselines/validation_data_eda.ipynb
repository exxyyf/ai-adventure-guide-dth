## Imports

```python
import os

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
VALIDATION_DATA_SHORT_PATH = "tourism_travel_questions_with_ground_truth_ver_short.csv"
VALIDATION_DATA_LONG_PATH = "tourism_travel_questions_with_ground_truth_ver_long.csv"
```

```python
PLOTS_DIR = "plots"
if os.path.exists(PLOTS_DIR):
    pass
else:
    os.mkdir(PLOTS_DIR)
```

```python
sns.set_style("whitegrid")
```

## EDA

```python
df = pd.read_csv(VALIDATION_DATA_SHORT_PATH)
```

```python
df.info()
```

### Question Type

```python
plt.figure(figsize=(7, 4))
sns.countplot(
    data=df,
    x=df["question_type"],
    hue=df["question_type"],
    palette="bright"
)
plt.savefig(
    os.path.join(
        PLOTS_DIR,
        "question_type_cnt.png",
    )
)
plt.show()
```

### Topic

```python
plt.figure(figsize=(10, 5))
sns.countplot(
    data=df,
    x=df["topic"],
    hue=df["topic"],
    palette="bright"
)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(
    os.path.join(
        PLOTS_DIR,
        "topic_cnt.png",
    )
)
plt.show()
```

### Difficulty Level

```python
plt.figure(figsize=(7, 5))
sns.countplot(
    data=df,
    x=df["difficulty_level"],
    hue=df["difficulty_level"],
    palette="bright"
)
plt.xticks(rotation=45)
plt.savefig(
    os.path.join(
        PLOTS_DIR,
        "difficulty_level_cnt.png",
    )
)
plt.show()
```
